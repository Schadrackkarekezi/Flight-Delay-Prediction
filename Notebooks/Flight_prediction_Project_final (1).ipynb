##DATA CLEANING

```python
#copy to avoid modifying the original
df_clean = df.copy()
```

### Check for duplicate rows

```python
duplicates = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")
if duplicates > 0:
    df_clean = df_clean.drop_duplicates()
    print(f"Removed {duplicates} duplicate rows")
```

### Create target variable for flight delays

```python
if 'DEP_DEL15' in df_clean.columns and 'ARR_DEL15' in df_clean.columns:
    df_clean['IS_DELAYED'] = ((df_clean['DEP_DEL15'] > 0) | (df_clean['ARR_DEL15'] > 0)).astype(int)
    print("\nCreated target variable 'IS_DELAYED'")
    print(f"Delay distribution: {df_clean['IS_DELAYED'].value_counts(normalize=True) * 100}")
```

### Handle missing values

```python
# For numerical columns: impute with median
numerical_columns = df_clean.select_dtypes(include=['int64', 'float64']).columns
for col in numerical_columns:
    missing_count = df_clean[col].isnull().sum()
    if missing_count > 0:
        median_value = df_clean[col].median()
        df_clean[col] = df_clean[col].fillna(median_value)
        print(f"Filled {missing_count} missing values in {col} with median ({median_value})")
```

# Handle time data: convert HHMM format to minutes since midnight

```python
time_columns = ['DEP_TIME', 'ARR_TIME']
for col in time_columns:
    if col in df_clean.columns:
        # Fill any missing values first
        if df_clean[col].isnull().sum() > 0:
            median_time = df_clean[col].median()
            df_clean[col] = df_clean[col].fillna(median_time)

        # Convert to minutes since midnight
        df_clean[f'{col}_MINUTES'] = (df_clean[col] // 100) * 60 + (df_clean[col] % 100)
        print(f"Created {col}_MINUTES feature from {col}")
```

# Drop columns not needed for modeling

```python
columns_to_drop = [
    "Unnamed: 21",
    "ORIGIN_AIRPORT_SEQ_ID",
    "DEST_AIRPORT_SEQ_ID",
    "OP_CARRIER_AIRLINE_ID",
    "OP_UNIQUE_CARRIER",
    "TAIL_NUM",
    "OP_CARRIER_FL_NUM",
    "CANCELLED",
    "DIVERTED"
]
# Only drop columns that exist in the DataFrame
columns_to_drop = [col for col in columns_to_drop if col in df_clean.columns]
if columns_to_drop:
    df_clean = df_clean.drop(columns=columns_to_drop)
    print(f"Dropped columns: {columns_to_drop}")
```

# Feature Engineering

```python
# Create cyclical features for time and date
# Day of week (1-7): convert to sine and cosine to represent cyclical nature
if 'DAY_OF_WEEK' in df_clean.columns:
    df_clean['DAY_OF_WEEK_SIN'] = np.sin(2 * np.pi * df_clean['DAY_OF_WEEK'] / 7)
    df_clean['DAY_OF_WEEK_COS'] = np.cos(2 * np.pi * df_clean['DAY_OF_WEEK'] / 7)
    print("Created cyclical features for DAY_OF_WEEK")
```

```python
# Day of month (1-31): convert to sine and cosine
if 'DAY_OF_MONTH' in df_clean.columns:
    df_clean['DAY_OF_MONTH_SIN'] = np.sin(2 * np.pi * df_clean['DAY_OF_MONTH'] / 31)
    df_clean['DAY_OF_MONTH_COS'] = np.cos(2 * np.pi * df_clean['DAY_OF_MONTH'] / 31)
    print("Created cyclical features for DAY_OF_MONTH")
```

```python
# Departure time: convert to sine and cosine to represent time of day
if 'DEP_TIME_MINUTES' in df_clean.columns:
    # Minutes in a day = 24 * 60 = 1440
    df_clean['DEP_TIME_SIN'] = np.sin(2 * np.pi * df_clean['DEP_TIME_MINUTES'] / 1440)
    df_clean['DEP_TIME_COS'] = np.cos(2 * np.pi * df_clean['DEP_TIME_MINUTES'] / 1440)
    print("Created cyclical features for DEP_TIME_MINUTES")
```

# Create time period categories

```python
# 2. Create time period categories
if 'DEP_TIME_MINUTES' in df_clean.columns:
    # Define time periods (in minutes since midnight)
    time_periods = [
        (0, 360, 'Early Morning'),    # 00:00-06:00
        (360, 600, 'Morning'),        # 06:00-10:00
        (600, 720, 'Mid-Day'),        # 10:00-12:00
        (720, 960, 'Afternoon'),      # 12:00-16:00
        (960, 1140, 'Evening'),       # 16:00-19:00
        (1140, 1440, 'Night')         # 19:00-24:00
    ]

    # Create time period feature
    def get_time_period(minutes):
        for start, end, label in time_periods:
            if start <= minutes < end:
                return label
        return 'Unknown'

    df_clean['DEP_TIME_PERIOD'] = df_clean['DEP_TIME_MINUTES'].apply(get_time_period)
    print("Created DEP_TIME_PERIOD feature")
```

#Create route features

```python
if 'ORIGIN' in df_clean.columns and 'DEST' in df_clean.columns:
    # Create route (combined origin and destination)
    df_clean['ROUTE'] = df_clean['ORIGIN'] + '_' + df_clean['DEST']
    print("Created ROUTE feature")

    # Get top routes and categorize less frequent routes as 'OTHER'
    route_counts = df_clean['ROUTE'].value_counts()
    top_routes = route_counts[route_counts >= 10].index  # Routes with at least 10 flights
    df_clean['ROUTE_CAT'] = df_clean['ROUTE'].apply(
        lambda x: x if x in top_routes else 'OTHER'
    )
    print(f"Created ROUTE_CAT feature with {len(top_routes)} categories + OTHER")
```

# 4. Create airport busyness features

```python
if 'ORIGIN' in df_clean.columns:
    # Count flights per origin airport and map back to create a feature
    origin_counts = df_clean['ORIGIN'].value_counts()
    df_clean['ORIGIN_FLIGHTS_COUNT'] = df_clean['ORIGIN'].map(origin_counts)
    print("Created ORIGIN_FLIGHTS_COUNT feature")

if 'DEST' in df_clean.columns:
    # Count flights per destination airport and map back to create a feature
    dest_counts = df_clean['DEST'].value_counts()
    df_clean['DEST_FLIGHTS_COUNT'] = df_clean['DEST'].map(dest_counts)
    print("Created DEST_FLIGHTS_COUNT feature")
```

# 5. Create distance features

```python
if 'DISTANCE' in df_clean.columns:
    # Create distance bins
    df_clean['DISTANCE_BIN'] = pd.cut(
        df_clean['DISTANCE'],
        bins=[0, 250, 500, 1000, 1500, 2000, 3000, float('inf')],
        labels=['Very Short', 'Short', 'Medium Short', 'Medium', 'Medium Long', 'Long', 'Very Long']
    )
    print("Created DISTANCE_BIN feature")
```

# 6. Create weekend/weekday feature

```python
if 'DAY_OF_WEEK' in df_clean.columns:
    # 6 and 7 are typically Saturday and Sunday
    df_clean['IS_WEEKEND'] = df_clean['DAY_OF_WEEK'].isin([6, 7]).astype(int)
    print("Created IS_WEEKEND feature")
```

# 7. Create rush hour feature

```python
if 'DEP_TIME_MINUTES' in df_clean.columns:
    # Define morning and evening rush hours
    morning_rush = (df_clean['DEP_TIME_MINUTES'] >= 420) & (df_clean['DEP_TIME_MINUTES'] <= 540)  # 7-9 AM
    evening_rush = (df_clean['DEP_TIME_MINUTES'] >= 960) & (df_clean['DEP_TIME_MINUTES'] <= 1140)  # 4-7 PM
    df_clean['IS_RUSH_HOUR'] = (morning_rush | evening_rush).astype(int)
    print("Created IS_RUSH_HOUR feature")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score

# --- Step 1: Balance the dataset ---
df_majority = df_clean[df_clean['IS_DELAYED'] == 0]
df_minority = df_clean[df_clean['IS_DELAYED'] == 1]

df_majority_downsampled = df_majority.sample(n=len(df_minority), random_state=42)
df_balanced = pd.concat([df_majority_downsampled, df_minority])
df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

# --- Step 2: Feature selection ---
x = df_balanced.drop([
    'IS_DELAYED', 'DEP_DEL15', 'ARR_DEL15', 'DEP_TIME', 'DEP_TIME_MINUTES',
    'ARR_TIME', 'ARR_TIME_MINUTES', 'DAY_OF_MONTH', 'DAY_OF_WEEK', 'ROUTE',
    'DISTANCE_BIN', 'AIRLINE_NAME', 'DEP_TIME_PERIOD',
    'ORIGIN_AIRPORT_ID', 'DEST_AIRPORT_ID'
], axis=1)

y = df_balanced['IS_DELAYED']

# --- Step 3: Train-test split ---
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15, random_state=42)

# Sample a subset from the training data
x_sample = x_train.sample(n=100_000, random_state=42)
y_sample = y_train.loc[x_sample.index]

# --- Step 4: Preprocessing ---
num_features = x.select_dtypes(include=["int64", "float64"]).columns
cat_features = x.select_dtypes(include=["object", "category"]).columns

numeric_transformer = StandardScaler()
oh_transformer = OneHotEncoder(drop='first', handle_unknown='ignore')

preprocessor = ColumnTransformer(
    transformers=[
        ("OneHotEncoder", oh_transformer, cat_features),
        ("StandardScaler", numeric_transformer, num_features)
    ]
)

# Fit on full training set and transform all needed datasets
preprocessor.fit(x_train)

x_sample_processed = preprocessor.transform(x_sample)
x_test_processed = preprocessor.transform(x_test)

# --- Step 5: Train individual models ---
log_clf = LogisticRegression(max_iter=1000, random_state=42)
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)
xgb_clf = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)

log_clf.fit(x_sample_processed, y_sample)
rf_clf.fit(x_sample_processed, y_sample)
xgb_clf.fit(x_sample_processed, y_sample)

# --- Step 6: Evaluate individual models ---
models = {
    "Logistic Regression": log_clf,
    "Random Forest": rf_clf,
    "XGBoost": xgb_clf
}

print("---- Individual Model Performance ----\n")
for name, model in models.items():
    y_pred = model.predict(x_test_processed)
    acc = accuracy_score(y_test, y_pred)
    print(f"{name} Accuracy: {acc:.4f}")
    print(classification_report(y_test, y_pred))
    if hasattr(model, "predict_proba"):
        auc = roc_auc_score(y_test, model.predict_proba(x_test_processed)[:, 1])
        print(f"{name} AUC: {auc:.4f}")
    print("-" * 50)

# --- Step 7: Hard Voting ---
hard_voting = VotingClassifier(estimators=[
    ('lr', log_clf),
    ('rf', rf_clf),
    ('xgb', xgb_clf)
], voting='hard')

hard_voting.fit(x_sample_processed, y_sample)
hard_pred = hard_voting.predict(x_test_processed)

print("\n---- Hard Voting Performance ----")
print("Accuracy:", accuracy_score(y_test, hard_pred))
print(classification_report(y_test, hard_pred))

# --- Step 8: Soft Voting ---
soft_voting = VotingClassifier(estimators=[
    ('lr', log_clf),
    ('rf', rf_clf),
    ('xgb', xgb_clf)
], voting='soft')

soft_voting.fit(x_sample_processed, y_sample)
soft_pred = soft_voting.predict(x_test_processed)

print("\n---- Soft Voting Performance ----")
print("Accuracy:", accuracy_score(y_test, soft_pred))
print(classification_report(y_test, soft_pred))
print("AUC:", roc_auc_score(y_test, soft_voting.predict_proba(x_test_processed)[:, 1]))

```

```python
from sklearn.model_selection import RandomizedSearchCV

rf_params = {
    'n_estimators': [100, 150],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5],
    'max_features': ['sqrt']
}

rf_search = RandomizedSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_distributions=rf_params,
    n_iter=8,
    cv=3,
    scoring='roc_auc',
    verbose=1,
    n_jobs=-1,
    random_state=42
)

rf_search.fit(x_sample_processed, y_sample)
best_rf = rf_search.best_estimator_
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

def save_and_show_conf_matrix(model, name, x_data, y_true, filename):
    # Get predictions
    y_pred = model.predict(x_data)
    cm = confusion_matrix(y_true, y_pred)

    # Set up figure
    fig, ax = plt.subplots(figsize=(6, 5))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["On Time (0)", "Delayed (1)"])
    disp.plot(ax=ax, cmap='Blues', values_format='d', colorbar=False)

    # ✅ Remove gridlines, ticks, and borders
    ax.grid(False)
    ax.set_xticks([])
    ax.set_yticks([])
    for spine in ax.spines.values():
        spine.set_visible(False)

    # Title and layout
    ax.set_title(f"{name} Confusion Matrix")
    plt.tight_layout()

    # Show and save
    plt.show()
    plt.savefig(filename, dpi=300)
    plt.close()

# Generate and save both confusion matrices
save_and_show_conf_matrix(hard_voting, "Hard Voting", x_test_processed, y_test, "hard_voting_confusion_matrix.png")
save_and_show_conf_matrix(soft_voting, "Soft Voting", x_test_processed, y_test, "soft_voting_confusion_matrix.png")
```
